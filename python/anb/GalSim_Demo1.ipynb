```python
import sys
import os
import math
import logging
import galsim
```

```python
def main(argv):
    """
    About as simple as it gets:
      - Use a circular Gaussian profile for the galaxy.
      - Convolve it by a circular Gaussian PSF.
      - Add Gaussian noise to the image.
    """
    # In non-script code, use getLogger(__name__) at module scope instead.
    logging.basicConfig(format="%(message)s", level=logging.INFO, stream=sys.stdout)
    logger = logging.getLogger("demo1") 

    gal_flux = 1.e5    # total counts on the image
    gal_sigma = 2.     # arcsec
    psf_sigma = 1.     # arcsec
    pixel_scale = 0.2  # arcsec / pixel
    noise = 30.        # standard deviation of the counts in each pixel

    logger.info('Starting demo script 1 using:')
    logger.info('    - circular Gaussian galaxy (flux = %.1e, sigma = %.1f),',gal_flux,gal_sigma)
    logger.info('    - circular Gaussian PSF (sigma = %.1f),',psf_sigma)
    logger.info('    - pixel scale = %.2f,',pixel_scale)
    logger.info('    - Gaussian noise (sigma = %.2f).',noise)

    # Define the galaxy profile
    gal = galsim.Gaussian(flux=gal_flux, sigma=gal_sigma)
    logger.debug('Made galaxy profile')

    # Define the PSF profile
    psf = galsim.Gaussian(flux=1., sigma=psf_sigma) # PSF flux should always = 1
    logger.debug('Made PSF profile')

    # Final profile is the convolution of these
    # Can include any number of things in the list, all of which are convolved 
    # together to make the final flux profile.
    final = galsim.Convolve([gal, psf])
    logger.debug('Convolved components into final profile')

    # Draw the image with a particular pixel scale, given in arcsec/pixel.
    # The returned image has a member, added_flux, which is gives the total flux actually added to 
    # the image.  One could use this value to check if the image is large enough for some desired
    # accuracy level.  Here, we just ignore it.
    image = final.drawImage(scale=pixel_scale)
    logger.debug('Made image of the profile: flux = %f, added_flux = %f',gal_flux,image.added_flux)

    # Add Gaussian noise to the image with specified sigma
    image.addNoise(galsim.GaussianNoise(sigma=noise))
    logger.debug('Added Gaussian noise')

    # Write the image to a file
    if not os.path.isdir('output'):
        os.mkdir('output')
    file_name = os.path.join('output','demo1.fits')
    # Note: if the file already exists, this will overwrite it.
    image.write(file_name)
    logger.info('Wrote image to %r' % file_name)  # using %r adds quotes around filename for us

    results = image.FindAdaptiveMom()

    logger.info('HSM reports that the image has observed shape and size:')
    logger.info('    e1 = %.3f, e2 = %.3f, sigma = %.3f (pixels)', results.observed_shape.e1,
                results.observed_shape.e2, results.moments_sigma)
    logger.info('Expected values in the limit that pixel response and noise are negligible:')
    logger.info('    e1 = %.3f, e2 = %.3f, sigma = %.3f', 0.0, 0.0, 
                math.sqrt(gal_sigma**2 + psf_sigma**2)/pixel_scale) 

if __name__ == "__main__":
    main(sys.argv)
```

### Now set the pixel scale to 0.387 arcsec/pixel and vary the flux.

#### I got this pixel scale from the header of lsc1m004-fl04-20141201-0123-e90_wcs_corrected.fits in the repository /data directory.

#### I am condensing the code to just the essential commands and looping through a list of flux values.

```python
import sys
import os
import math
import galsim
```

```python
def main(argv):
    
    # A list of fluxes to plot
    flux_list = [1.e8, 1.e7, 1.e6, 1.e5]
    
    # Output filenames
    output_filename = ['demo1_flux1.fits', 'demo1_flux2.fits', 'demo1_flux3.fits', 'demo1_flux4.fits']
    
    gal_sigma = 2.       # arcsec
    psf_sigma = 1.       # arcsec
    pixel_scale = 0.387  # arcsec / pixel
    noise = 30.          # standard deviation of the counts in each pixel
    
    filename_index = 0
    
    for gal_flux in flux_list:

        gal = galsim.Gaussian(flux=gal_flux, sigma=gal_sigma)
        psf = galsim.Gaussian(flux=1., sigma=psf_sigma) # PSF flux should always = 1
        final = galsim.Convolve([gal, psf])
        image = final.drawImage(scale=pixel_scale)
        image.addNoise(galsim.GaussianNoise(sigma=noise))
        if not os.path.isdir('output'):
            os.mkdir('output')
        file_name = os.path.join('output',output_filename[filename_index]) # <-- Write to different file
        # Note: if the file already exists, this will overwrite it.
        image.write(file_name)

        filename_index = filename_index + 1
    
if __name__ == "__main__":
    main(sys.argv)
```

### Next I will try Moffat PSF profiles using some of the data Andrew presented (with constant flux)

I calculate FWHM for the Moffat profiles with the formula: $\ FWHM = 2 \alpha \sqrt{2^{\frac{1}{\beta}} - 1}$

```python
import sys
import os
import math
import galsim
```

```python
def main(argv):
    
    # A list of alpha parameters for the Moffat profiles
    alpha_list = [3.08, 3.00, 4.87, 5.14]
    
    # A list of beta parameters for the Moffat profiles
    beta_list = [1.9, 2.42, 2.8, 3.08]
    
    # Output filenames
    output_filename = ['demo1_Moffat1.fits', 'demo1_Moffat2.fits',
                       'demo1_Moffat3.fits', 'demo1_Moffat4.fits']
    
    gal_flux = 1.e8
    gal_sigma = 5.       # arcsec
    psf_sigma = 1.       # arcsec
    pixel_scale = 0.387  # arcsec / pixel
    noise = 30.          # standard deviation of the counts in each pixel
    
    filename_index = 0
    alpha_index = 0
    
    for gal_beta in beta_list:

        # calculate fwhm from alpha and beta
        # fwhm is one of the possible size parameters in the GalSim Moffat function
        gal_fwhm = alpha_list[alpha_index] * (2 * math.sqrt((2 ** (1 / gal_beta)) - 1))
        
        gal = galsim.Gaussian(flux=gal_flux, sigma=gal_sigma)
        psf = galsim.Moffat(beta=gal_beta, fwhm=gal_fwhm, flux=1.) # PSF flux should always = 1
        final = galsim.Convolve([gal, psf])
        image = final.drawImage(scale=pixel_scale)
        image.addNoise(galsim.GaussianNoise(sigma=noise))
        if not os.path.isdir('output'):
            os.mkdir('output')
        file_name = os.path.join('output',output_filename[filename_index]) # <-- Write to different file
        # Note: if the file already exists, this will overwrite it.
        image.write(file_name)
        
        filename_index = filename_index + 1
        alpha_index = alpha_index + 1
    
if __name__ == "__main__":
    main(sys.argv)
```

### Vary the rotation

```python
def main(argv):
    
    # to calculate FWHM for the Moffat profile
    gal_alpha = 4.87
    gal_beta = 2.80
    
    # calculate fwhm from alpha and beta
    # fwhm is one of the possible size parameters in the GalSim Moffat function
    gal_fwhm = gal_alpha * (2 * math.sqrt((2 ** (1 / gal_beta)) - 1))
    
    # Output filenames
    output_filename = ['demo1_Sersic13.fits', 'demo1_Sersic14.fits',
                       'demo1_Sersic15.fits', 'demo1_Sersic16.fits']
    
    # List of g1 values for galaxy shear
    rotation_list = [23.0, 45.0, 68.0, 90.0]
    
    sersic_g1 = 0.4
    sersic_g2 = 0.05
    sersic_index = 1
    gal_re = 5.0           # half light radius
    gal_flux = 1.e8
    gal_sigma = 5.         # arcsec
    psf_sigma = 1.         # arcsec
    pixel_scale = 0.387    # arcsec / pixel
    noise = 30.            # standard deviation of the counts in each pixel
    
    filename_index = 0
    
    for sersic_rotation in rotation_list:
        
        gal = galsim.Sersic(n=sersic_index, half_light_radius=gal_re, flux=gal_flux)
        gal = gal.shear(g1=sersic_g1, g2=sersic_g2)
        gal.applyRotation(sersic_rotation*galsim.degrees)
        psf = galsim.Moffat(beta=gal_beta, fwhm=gal_fwhm, flux=1.) # PSF flux should always = 1
        final = galsim.Convolve([gal, psf])
        image = final.drawImage(scale=pixel_scale)
        image.addNoise(galsim.GaussianNoise(sigma=noise))
        if not os.path.isdir('output'):
            os.mkdir('output')
        file_name = os.path.join('output',output_filename[filename_index]) # <-- Write to different file
        # Note: if the file already exists, this will overwrite it.
        image.write(file_name)
        filename_index = filename_index + 1
    
if __name__ == "__main__":
    main(sys.argv)
```
