```python
import sys
import os
import math
import numpy as np
import pyfits
import matplotlib.pyplot as plt
import time
import galsim
```

```python
gal_flux = 1.e5      # total counts on the image
gal_sigma = 2.       # arcsec
psf_sigma = 1.       # arcsec
pixel_scale = 0.387  # arcsec / pixel
xsize = 100
ysize = 100
noise = 30.
dx = 3
dy = 3
```

```python
big_image = galsim.ImageF(xsize, ysize)
```

```python
b1 = galsim.BoundsI(23,48,23,48)
sub_image_1 = big_image[b1]
gal_1 = galsim.Gaussian(flux=gal_flux, sigma=gal_sigma)
#psf_1 = galsim.Gaussian(flux=1., sigma=psf_sigma)
#final_1 = galsim.Convolve([gal_1, psf_1])
```

```python
gal_1.drawImage(sub_image_1, scale=pixel_scale)
```

```python
b2 = galsim.BoundsI(38,58,38,58)
sub_image_2 = big_image[b2]
```

```python
gal_2 = galsim.Gaussian(flux=gal_flux, sigma=gal_sigma)
#psf_2 = galsim.Gaussian(flux=1., sigma=psf_sigma)
#final_2 = galsim.Convolve([gal_2, psf_2])
gal_2.drawImage(sub_image_2, scale=pixel_scale)
```

```python
big_image.write("Aaron_test.fits")
hdus = pyfits.open("Aaron_test.fits")
data = hdus[0].data
fig = plt.figure(figsize=(8, 8))
im = plt.imshow(data, cmap="gray")
```

```python
big_image.addNoise(galsim.GaussianNoise(sigma=noise))
```

```python
gal_1 = galsim.Gaussian(flux=gal_flux, sigma=gal_sigma)
gal_2 = galsim.Gaussian(flux=gal_flux, sigma=gal_sigma)
gal_2 = gal_2.shift(dx,dy)
#psf11 = galsim.Gaussian(flux=1., sigma=psf_sigma)
#psf12 = galsim.Gaussian(flux=1., sigma=psf_sigma)
#psf12 = psf12.shift(dx,dy)
#psf1 = psf11 + psf12
final = gal_1 + gal_2
final.drawImage(big_image, scale=pixel_scale)
```
