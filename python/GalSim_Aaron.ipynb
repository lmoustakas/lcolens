```python
import sys
import os
import math
import numpy as np
import pyfits
import matplotlib.pyplot as plt
import time
import galsim
```

```python
gal_flux = 1.e5      # total counts on the image
gal_sigma = 2.       # arcsec
psf_sigma = 1.       # arcsec
pixel_scale = 0.387  # arcsec / pixel
xsize = 128
ysize = 128
noise = 30.
dx = 10
dy = 5
```

```python
big_image = galsim.ImageF(xsize, ysize)
```

```python
gal11 = galsim.Gaussian(flux=gal_flux, sigma=gal_sigma)
gal12 = galsim.Gaussian(flux=gal_flux, sigma=gal_sigma)
gal12 = gal12.shift(dx,dy)
gal1 = gal11 + gal12
psf11 = galsim.Gaussian(flux=1., sigma=psf_sigma)
psf12 = galsim.Gaussian(flux=1., sigma=psf_sigma)
psf12 = psf12.shift(dx,dy)
psf1 = psf11 + psf12
final1 = galsim.Convolve([gal1, psf1])
```

```python
b1 = galsim.BoundsI(25, 95, 5, 75)
```

```python
sub_image1 = big_image[b1]
```

```python
final1.drawImage(sub_image1, scale=pixel_scale)
```

```python
gal2 = galsim.Gaussian(flux=gal_flux, sigma=gal_sigma)
psf2 = galsim.Gaussian(flux=1., sigma=psf_sigma)
final2 = galsim.Convolve([gal2, psf2])
```

```python
b2 = galsim.BoundsI(60, 90, 10, 40)
```

```python
sub_image2 = big_image[b2]
```

```python
final2.drawImage(sub_image2, scale=pixel_scale)
```

```python
big_image.addNoise(galsim.GaussianNoise(sigma=noise))
```

```python
big_image.write("Aaron_test.fits")
```
