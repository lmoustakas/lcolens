```python
def quad_image_model(theta, N_pix):
    x0, y0, amplitude0, \
    x1, y1, amplitude1, \
    x2, y2, amplitude2, \
    x3, y3, amplitude3, \
    alpha, beta, N_bkg = theta

    xg, yg = np.mgrid[:N_pix,:N_pix]

    #print x0,y0, amplitude0
    p0  =  twoD_Moffat((xg, yg), amplitude0, alpha, beta, x0, y0, 0)
    p1  =  twoD_Moffat((xg, yg), amplitude1, alpha, beta, x1, y1, 0)
    p2  =  twoD_Moffat((xg, yg), amplitude2, alpha, beta, x2, y2, 0)
    p3  =  twoD_Moffat((xg, yg), amplitude3, alpha, beta, x3, y3, 0)
    model = (p0+p1+p2+p3).reshape(N_pix, N_pix)
    model+=N_bkg
    #print model
    #model +=  twoD_Moffat((xg, yg), amplitude1, alpha, beta, x1, y1, N_bkg)
    #model +=  twoD_Moffat((xg, yg), amplitude2, alpha, beta, x2, y2, N_bkg)
    #model +=  twoD_Moffat((xg, yg), amplitude3, alpha, beta, x3, y3, N_bkg)
    return model
```

```python
# AN EXAMPLE OF A QUAD LENS IMAGE MODEL
x0,y0,amp0 = 12,15,1200
x1,y1,amp1 = 18,15,900
x2,y2,amp2 = 15.5,12,800
x3,y3,amp3 = 15,18,700
theta = [y0,x0,amp0, y1,x1,amp1, y2,x2,amp2, y3,x3,amp3, 3., 3., 300]
model = quad_image_model(theta, N_pixels)
plt.imshow(model, cmap='gray', interpolation='none')
plt.colorbar()
plt.show()
```

```python
def moffat_chi(theta, image):
    model = quad_image_model(theta, len(image))
    chi = (image - model)/np.sqrt(image)

    return chi
```

```python
# EMCEE WANTS A PRIOR DISTRIBUTUION AND A LIKELIHOOD

def lnprior(_theta):
    _x0, _y0, _amplitude0, \
    _x1, _y1, _amplitude1, \
    _x2, _y2, _amplitude2, \
    _x3, _y3, _amplitude3, \
    _alpha, _beta, _N_bkg = _theta

    
    x0, y0, amplitude0, \
    x1, y1, amplitude1, \
    x2, y2, amplitude2, \
    x3, y3, amplitude3, \
    alpha, beta, N_bkg = theta_test

    p=3.
    if  x0-p <= _x0 <= x0+p \
	and y0-p <= _y0 <= y0+p \
	and x1-p <= _x1 <= x1+p \
	and y1-p <= _y1 <= y1+p \
	and x2-p <= _x2 <= x2+p \
	and y2-p <= _y2 <= y2+p \
	and x3-p <= _x3 <= x3+p \
	and y3-p <= _y3 <= y3+p \
    and 0. <= _amplitude0 <= 100.*amplitude0 \
    and 0. <= _amplitude1 <= 100.*amplitude1 \
    and 0. <= _amplitude2 <= 100.*amplitude2 \
    and 0. <= _amplitude3 <= 100.*amplitude3 \
	and 1. < _alpha < 3. \
	and 1. < _beta < 5. \
	and 0. < _N_bkg < np.max(obj.image):
	  return 0.0  #sigma and tau are scale parameters. Also true of delta_mag and avg_mag, but they are already in logarithmic units.
    else:
        #print 'fing up'
        return -np.inf
    
def moffat_ll(theta, image):
    ll = sum(np.ravel(moffat_chi(theta, image))**2) 
    return ll
    
    
def lnprob(theta, image):
    lp = lnprior(theta)
    #print lp, theta
    if not np.isfinite(lp):
        return -np.inf
    mll = moffat_ll(theta, image)
    #print mll
    if not np.isfinite(mll):
        return -np.inf
    return lp + mll

```
