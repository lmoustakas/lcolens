```python
plt.rcParams['figure.figsize']=(10,8)
```

```python
Vizier.ROW_LIMIT = -1
```

```python
v=Vizier(columns=['_RAJ2000','_DEJ2000','B-V','Vmag'],column_filters={"Vmag":">5"})
```

```python
v.ROW_LIMIT=-1
```

```python
result = v.query_region("HE0435-1223", width="40m", catalog='NOMAD1')
```

```python
print result
```

```python
ra,dec=result[0]['_RAJ2000'],result[0]['_DEJ2000']
#print(result[0]['_RAJ2000','_DEJ2000','Vmag'])
```

```python
area=10*(18-result[0]['Vmag'])
```

```python
plt.scatter(result[0]['_RAJ2000'],result[0]['_DEJ2000'],s=area,marker='o')
ax=plt.gca()
ax.set_xlim(ax.get_xlim()[::-1])
```

```python
hdulist=fits.open('../data/HE0435_LCOGT/lsc1m004-fl04-20141201-0123-e90.fits')
w = wcs.WCS(hdulist[0].header)
hdulist.close()
#print w.wcs.cd
```

```python
x,y=w.wcs_world2pix(ra,dec,1)
```

```python
plt.scatter(x,y,s=area,marker='o')
```

```python
x,y=w.all_world2pix(ra,dec,1,tolerance=4e3)
```
