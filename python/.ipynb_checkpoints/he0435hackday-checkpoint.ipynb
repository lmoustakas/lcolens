```python
plt.rcParams['figure.figsize']=(10,8)
```

```python
Vizier.ROW_LIMIT = -1
```

```python
v=Vizier(columns=['_RAJ2000','_DEJ2000','B-V','Vmag'],column_filters={"Vmag":">5"})
```

```python
v.ROW_LIMIT=-1
```

```python
result = v.query_region("HE0435-1223", width="45m", catalog='NOMAD1')
```

```python
print result
```

```python
ra,dec=result[0]['_RAJ2000'],result[0]['_DEJ2000']
#print(result[0]['_RAJ2000','_DEJ2000','Vmag'])
```

```python
area=10*(18-result[0]['Vmag'])
```

```python
# This works but is inelegant
#plt.scatter(result[0]['_RAJ2000'],result[0]['_DEJ2000'],s=area,marker='o')
#ax=plt.gca()
#ax.set_xlim(ax.get_xlim()[::-1])
```

```python
hdulist=fits.open('../data/HE0435_LCOGT/lsc1m004-fl04-20141201-0123-e90.fits')
```

```python
#w = wcs.WCS(hdulist[0].header)
hdulist.close()
#print w.wcs.cd
```

```python
bigw=WCS(hdulist[0].header)
```

```python
fig=plt.figure()
ax=fig.add_axes([0.15,0.1,0.8,0.8],projection=bigw)
ax.scatter(ra,dec,s=area,marker='o')
```

```python
# Let's convert our NOMAD stellar list to x,y pixel coordinates
x,y=bigw.wcs_world2pix(ra,dec,1)
```

```python
fig=plt.figure()
ax=fig.add_axes([0.15,0.1,0.8,0.8])
ax.scatter(x,y,s=area,marker='o')
#ax=fig.add_subplot(111)
ax.set_xlim(-200,4200)
ax.set_ylim(-200,4200)
ax.set_aspect(1.0)
```
