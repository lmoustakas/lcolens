```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
from astropy import wcs
from wcsaxes import WCS
import aplpy
import time
import os
import glob
import os.path
```

# Insert all lists needed to store data, initialize them with [] to create an empty list.

```python
#MJD
lsc_mjd, cpt_mjd, coj_mjd = [], [], []
#Filters
lsc_filter, cpt_filter, coj_filter = [], [], []
#Sky Brightness
lsc_meas_bright, cpt_meas_bright, coj_meas_bright = [], [], []
lsc_calc_bright, cpt_calc_bright, coj_calc_bright = [], [], []
lsc_dev_bright, cpt_dev_bright, coj_dev_bright = [], [], []
#Mean Airmass
lsc_mean_airmass, cpt_mean_airmass, coj_mean_airmass = [], [], []
#Brightness with respect to filter used
lsc_rp178_bright, lsc_rp179_bright, lsc_gp179_bright, lsc_gp178_bright = [], [], [], []
cpt_gp181_bright, cpt_rp181_bright = [], []
coj_gp184_bright, coj_rp184_bright, coj_rp183_bright = [], [], []
#MJD with respect to filter used
lsc_rp178_mjd, lsc_rp179_mjd, lsc_gp179_mjd, lsc_gp178_mjd = [], [], [], []
cpt_gp181_mjd, cpt_rp181_mjd = [], []
coj_gp184_mjd, coj_rp184_mjd, coj_rp183_mjd = [], [], []
#Gain
all_gain, lsc_gain, cpt_gain, coj_gain = [], [], [], []
```

# Iterate over FITS files to obtain data.

```python
#topdir = '../'
#exten = '.fits'
 
def step(ext, dirname, names):
    ext = ext.lower()
 
    for name in names:
        if name.lower().endswith('fits'):
            fits_file_name = (os.path.join(dirname, name))
            hdulist=fits.open(fits_file_name)
            site_id = hdulist[0].header['SITEID']
            gain = hdulist[0].header['GAIN']
            all_gain.append(gain)
            if site_id == 'lsc':
                lsc_meas_bright.append(hdulist[0].header['WMSSKYBR'])
                lsc_calc_bright.append(hdulist[0].header['SKYMAG'])
                lsc_dev_bright.append(hdulist[0].header['WMSSKYBR'] - hdulist[0].header['SKYMAG'])
                lsc_mjd.append(hdulist[0].header['MJD-OBS'])
                lsc_mean_airmass.append(hdulist[0].header['AIRMASS'])
                lsc_filter.append(hdulist[0].header['FILTERI2'])
                lsc_gain.append(gain)
                if hdulist[0].header['FILTERI2'] == 'SDSS-RP-178':
                    lsc_rp178_bright.append(hdulist[0].header['WMSSKYBR'])
                    lsc_rp178_mjd.append(hdulist[0].header['MJD-OBS'])
                elif hdulist[0].header['FILTERI2'] == 'SDSS-RP-179':
                    lsc_rp179_bright.append(hdulist[0].header['WMSSKYBR']) 
                    lsc_rp179_mjd.append(hdulist[0].header['MJD-OBS'])
                elif hdulist[0].header['FILTERI2'] == 'SDSS-GP-179':
                    lsc_gp179_bright.append(hdulist[0].header['WMSSKYBR'])
                    lsc_gp179_mjd.append(hdulist[0].header['MJD-OBS'])
                elif hdulist[0].header['FILTERI2'] == 'SDSS-GP-178':
                    lsc_gp178_bright.append(hdulist[0].header['WMSSKYBR']) 
                    lsc_gp178_mjd.append(hdulist[0].header['MJD-OBS'])
            elif site_id == 'cpt':
                cpt_meas_bright.append(hdulist[0].header['WMSSKYBR'])
                cpt_calc_bright.append(hdulist[0].header['SKYMAG'])
                cpt_dev_bright.append(hdulist[0].header['WMSSKYBR'] - hdulist[0].header['SKYMAG'])  
                cpt_mjd.append(hdulist[0].header['MJD-OBS'])
                cpt_mean_airmass.append(hdulist[0].header['AIRMASS'])
                cpt_filter.append(hdulist[0].header['FILTERI2'])
                cpt_gain.append(gain)
                if hdulist[0].header['FILTERI2'] == 'SDSS-GP-181':
                    cpt_gp181_bright.append(hdulist[0].header['WMSSKYBR'])
                    cpt_gp181_mjd.append(hdulist[0].header['MJD-OBS'])
                elif hdulist[0].header['FILTERI2'] == 'SDSS-RP-181':
                    cpt_rp181_bright.append(hdulist[0].header['WMSSKYBR']) 
                    cpt_rp181_mjd.append(hdulist[0].header['MJD-OBS'])
            elif site_id == 'coj':
                coj_meas_bright.append(hdulist[0].header['WMSSKYBR'])
                coj_calc_bright.append(hdulist[0].header['SKYMAG'])
                coj_dev_bright.append(hdulist[0].header['WMSSKYBR'] - hdulist[0].header['SKYMAG']) 
                coj_mjd.append(hdulist[0].header['MJD-OBS'])
                coj_mean_airmass.append(hdulist[0].header['AIRMASS'])
                coj_filter.append(hdulist[0].header['FILTERI2'])
                coj_gain.append(gain)
                if hdulist[0].header['FILTERI2'] == 'SDSS-GP-184':
                    coj_gp184_bright.append(hdulist[0].header['WMSSKYBR'])
                    coj_gp184_mjd.append(hdulist[0].header['MJD-OBS'])
                elif hdulist[0].header['FILTERI2'] == 'SDSS-RP-184':
                    coj_rp184_bright.append(hdulist[0].header['WMSSKYBR']) 
                    coj_rp184_mjd.append(hdulist[0].header['MJD-OBS'])
                elif hdulist[0].header['FILTERI2'] == 'SDSS-RP-183':
                    coj_rp183_bright.append(hdulist[0].header['WMSSKYBR'])
                    coj_rp183_mjd.append(hdulist[0].header['MJD-OBS'])
            else:
                print 'Check %s' % fits_file_name

os.path.walk('../../Data/', step, 'fits')
# If first argument is '../' it will iterate over 3 FITS files in Github repository.
# If first argument is '../../Data/' it will iterate over entire FITS file database, if this is where you have them stored.
```
